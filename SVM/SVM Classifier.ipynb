```python
# Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Load dataset
df = pd.read_csv('emails.csv')
print(df)
```

```python
# Features and target
X = df[['WordCount', 'NumLinks']]
y = df['Spam']
```

```python
# Scale features (SVM performs better with scaled data)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Create & train SVM
svm_model = SVC(kernel='linear') 
svm_model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = svm_model.predict(X_test)
```

```python
# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# Predict new email
new_email = scaler.transform([[250, 3]])  # Scale new input!
prediction = svm_model.predict(new_email)
print(f"\nIs an email with 250 words & 3 links spam? {'Yes' if prediction[0]==1 else 'No'}")
```
